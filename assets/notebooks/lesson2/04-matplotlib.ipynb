### äºŒã€ä¸¤ç§ç»˜å›¾æ¥å£ï¼šé¢å‘å¯¹è±¡ vs. Pyplot


Matplotlib åŒæ—¶æä¾›é¢å‘å¯¹è±¡ (OO) æ¥å£å’ŒåŸºäºçŠ¶æ€çš„ Pyplot æ¥å£ã€‚æœ¬æ•™ç¨‹æ¨è OO é£æ ¼ï¼Œå› ä¸ºåœ¨å¤æ‚å›¾è¡¨ä¸­æ›´æ¸…æ™°ã€æ›´å¯æ§ï¼Œä½†äº†è§£ Pyplot é£æ ¼æœ‰åŠ©äºé˜…è¯»ä»–äººä»£ç ã€‚

```python
x = np.linspace(0, 10, 100)


y = np.sin(x)




fig, ax = plt.subplots()


ax.plot(x, y)


ax.set_title("é¢å‘å¯¹è±¡ (OO) é£æ ¼")


plt.show()




plt.figure()


plt.plot(x, y)


plt.title("Pyplot é£æ ¼")


plt.show()
```

### ä¸‰ã€ç»˜åˆ¶å¸¸è§çš„å›¾è¡¨


Matplotlib æ”¯æŒå‡ ä¹æ‰€æœ‰å¸¸è§çš„åŸºç¡€å›¾è¡¨ç±»å‹ã€‚ä¸‹é¢æŒ‰ç…§åœºæ™¯æ¼”ç¤ºæŠ˜çº¿å›¾ã€æŸ±çŠ¶å›¾ã€ç›´æ–¹å›¾ã€æ•£ç‚¹å›¾ã€é¥¼å›¾å’Œç®±å½¢å›¾çš„ç»˜åˆ¶ã€‚

#### 1. æŠ˜çº¿å›¾ (`plot`)

æŠ˜çº¿å›¾é€‚åˆå±•ç¤ºè¿ç»­æ—¶é—´æˆ–é¡ºåºå˜é‡çš„å˜åŒ–è¶‹åŠ¿ã€‚

| å‚æ•° | è¯´æ˜ |
| --- | --- |
| `x`, `y` | è¦ç»˜åˆ¶çš„æ•°æ®åºåˆ—ã€‚ |
| `color` | çº¿æ¡é¢œè‰²ã€‚ |
| `linestyle` | çº¿å‹ï¼Œä¾‹å¦‚ `'-'`, `'--'`, `':'`ã€‚ |
| `linewidth` | çº¿å®½ã€‚ |
| `marker` | æ•°æ®ç‚¹æ ‡è®°æ ·å¼ã€‚ |
| `label` | å›¾ä¾‹æ ‡ç­¾ã€‚ |
| `alpha` | é€æ˜åº¦ã€‚ |

```python
x_data = np.linspace(0, 10, 100)

y_data = np.sin(x_data)

fig, ax = plt.subplots(figsize=(8, 4))

ax.plot(x_data, y_data, color="steelblue", linewidth=2, marker="o", markevery=10, label="sin(x)")
ax.set_title("ç®€å•çš„æ­£å¼¦å‡½æ•°æŠ˜çº¿å›¾")
ax.set_xlabel("X å€¼")
ax.set_ylabel("sin(X)")
ax.legend(loc="best")

plt.show()
```

#### 2. æŸ±çŠ¶å›¾ (`bar` ä¸ `barh`)

æŸ±çŠ¶å›¾ç”¨äºæ¯”è¾ƒä¸åŒç±»åˆ«ä¹‹é—´çš„æ•°å€¼å·®å¼‚ï¼Œç±»åˆ«åç§°è¾ƒé•¿æ—¶å¯ä»¥è€ƒè™‘æ°´å¹³æŸ±çŠ¶å›¾ã€‚

| æ–¹æ³• | å¸¸ç”¨å‚æ•° | è¯´æ˜ |
| --- | --- | --- |
| `Axes.bar(x, height, width=0.8, color=None, label=None, alpha=None)` | `x`: ç±»åˆ«ï¼›`height`: æ•°å€¼ï¼›`width`: æŸ±ä½“å®½åº¦ï¼›`color`: é¢œè‰²ï¼›`label`: å›¾ä¾‹æ ‡ç­¾ï¼›`alpha`: é€æ˜åº¦ã€‚ | å‚ç›´æŸ±çŠ¶å›¾ã€‚ |
| `Axes.barh(y, width, height=0.8, color=None, label=None, alpha=None)` | `y`: ç±»åˆ«ï¼›`width`: æ•°å€¼ï¼›å…¶ä½™å‚æ•°å«ä¹‰ç›¸åŒã€‚ | æ°´å¹³æŸ±çŠ¶å›¾ã€‚ |

```python
categories = ["ç¬¬ä¸€å­£åº¦", "ç¬¬äºŒå­£åº¦", "ç¬¬ä¸‰å­£åº¦", "ç¬¬å››å­£åº¦"]
values = [150, 230, 180, 210]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4), constrained_layout=True)

ax1.bar(categories, values, width=0.6, color=["#ff9999", "#66b3ff", "#99ff99", "#ffcc99"], edgecolor="black", label="å­£åº¦é”€å”®")
ax1.set_title("å‚ç›´æŸ±çŠ¶å›¾ (bar)")
ax1.set_ylabel("é”€å”®é¢ (ä¸‡å…ƒ)")
ax1.legend()

ax2.barh(categories, values, height=0.6, color=["#ff9999", "#66b3ff", "#99ff99", "#ffcc99"], label="å­£åº¦é”€å”®", alpha=0.8)
ax2.set_title("æ°´å¹³æŸ±çŠ¶å›¾ (barh)")
ax2.set_xlabel("é”€å”®é¢ (ä¸‡å…ƒ)")
ax2.legend(loc="lower right")

fig.suptitle("å¹´åº¦é”€å”®é¢å¯¹æ¯”")

plt.show()
```

#### 3. ç›´æ–¹å›¾ (`hist`)

ç›´æ–¹å›¾ç”¨äºæŸ¥çœ‹å•ä¸ªæ•°å€¼å‹å˜é‡çš„åˆ†å¸ƒæƒ…å†µã€‚

| å‚æ•° | è¯´æ˜ |
| --- | --- |
| `x` | è¾“å…¥æ•°æ®ã€‚ |
| `bins` | åˆ†ç®±çš„æ•°é‡æˆ–è¾¹ç•Œã€‚ |
| `range` | åˆ†ç®±èŒƒå›´ã€‚ |
| `density` | æ˜¯å¦æ˜¾ç¤ºæ¦‚ç‡å¯†åº¦ã€‚ |
| `color` | æŸ±ä½“é¢œè‰²ã€‚ |
| `alpha` | é€æ˜åº¦ã€‚ |
| `edgecolor` | æŸ±ä½“è¾¹æ¡†é¢œè‰²ã€‚ |

```python
data = np.random.randn(1000)

fig, ax = plt.subplots(figsize=(7, 4))
ax.hist(data, bins=30, range=(-4, 4), color="#66b3ff", alpha=0.8, edgecolor="black")
ax.set_title("éšæœºæ•°æ®çš„åˆ†å¸ƒç›´æ–¹å›¾")
ax.set_xlabel("æ•°å€¼")
ax.set_ylabel("é¢‘æ•°")

plt.show()
```

#### 4. æ•£ç‚¹å›¾ (`scatter`)

æ•£ç‚¹å›¾ç”¨äºè§‚å¯Ÿä¸¤ä¸ªæ•°å€¼å˜é‡ä¹‹é—´çš„å…³ç³»ã€‚

| å‚æ•° | è¯´æ˜ |
| --- | --- |
| `x`, `y` | ç‚¹çš„åæ ‡ã€‚ |
| `s` | ç‚¹çš„å¤§å°ã€‚ |
| `c` | ç‚¹çš„é¢œè‰²æˆ–æ•°å€¼åˆ—ã€‚ |
| `cmap` | å½“ `c` ä¸ºæ•°å€¼æ—¶ä½¿ç”¨çš„é¢œè‰²æ˜ å°„ã€‚ |
| `alpha` | é€æ˜åº¦ã€‚ |
| `edgecolors` | ç‚¹çš„è¾¹æ¡†é¢œè‰²ã€‚ |
| `label` | å›¾ä¾‹æ ‡ç­¾ã€‚ |

```python
x_scatter = np.random.rand(50) * 10
y_scatter = 2 * x_scatter + 1 + np.random.randn(50) * 2
sizes = (np.random.rand(50) * 80) + 20

fig, ax = plt.subplots(figsize=(6, 4))
ax.scatter(x_scatter, y_scatter, s=sizes, c=x_scatter, cmap="viridis", alpha=0.7, edgecolors="black", label="æ ·æœ¬ç‚¹")
ax.set_title("X å’Œ Y çš„æ•£ç‚¹å…³ç³»å›¾")
ax.set_xlabel("å˜é‡ X")
ax.set_ylabel("å˜é‡ Y")
ax.grid(True)
ax.legend(loc="upper left")

plt.show()
```

#### 5. é¥¼å›¾ (`pie`)

é¥¼å›¾ç”¨äºå±•ç¤ºå„éƒ¨åˆ†å æ•´ä½“çš„æ¯”ä¾‹ï¼Œä½†å½“ç±»åˆ«è¿‡å¤šæ—¶å¯è¯»æ€§è¾ƒå·®ã€‚

| å‚æ•° | è¯´æ˜ |
| --- | --- |
| `x` | æ¯ä¸ªæ‰‡åŒºçš„æ•°å€¼ã€‚ |
| `explode` | çªå‡ºæ˜¾ç¤ºæŸäº›æ‰‡åŒºçš„åç§»é‡ã€‚ |
| `labels` | æ‰‡åŒºæ ‡ç­¾ã€‚ |
| `autopct` | ç™¾åˆ†æ¯”æ ¼å¼å­—ç¬¦ä¸²ã€‚ |
| `startangle` | èµ·å§‹è§’åº¦ã€‚ |
| `shadow` | æ˜¯å¦æ·»åŠ é˜´å½±ã€‚ |
| `colors` | è‡ªå®šä¹‰é…è‰²ã€‚ |
| `pctdistance` | ç™¾åˆ†æ¯”æ–‡æœ¬è·ç¦»åœ†å¿ƒçš„æ¯”ä¾‹ã€‚ |

```python
labels = ["å¸‚åœºéƒ¨", "ç ”å‘éƒ¨", "é”€å”®éƒ¨", "è¡Œæ”¿éƒ¨"]
sizes = [15, 30, 45, 10]
explode = (0, 0, 0.1, 0)
colors = ["#ff9999", "#66b3ff", "#ffcc99", "#99ff99"]

fig, ax = plt.subplots()
ax.pie(
    sizes,
    explode=explode,
    labels=labels,
    colors=colors,
    autopct="%1.1f%%",
    shadow=True,
    startangle=90,
    pctdistance=0.8
)
ax.axis("equal")
ax.set_title("å…¬å¸å„éƒ¨é—¨äººæ•°å æ¯”")

plt.show()
```

#### 6. ç®±å½¢å›¾ (`boxplot`)

ç®±å½¢å›¾å±•ç¤ºæ•°æ®çš„ä¸­ä½æ•°ã€å››åˆ†ä½æ•°ä»¥åŠæ½œåœ¨çš„å¼‚å¸¸å€¼ã€‚

| å‚æ•° | è¯´æ˜ |
| --- | --- |
| `x` | è¦ç»˜åˆ¶çš„æ•°æ®åºåˆ—ï¼ˆæˆ–åºåˆ—åˆ—è¡¨ï¼‰ã€‚ |
| `notch` | æ˜¯å¦ç»˜åˆ¶å‡¹å£ä»¥è§‚å¯Ÿç½®ä¿¡åŒºé—´ã€‚ |
| `vert` | æ˜¯å¦å‚ç›´æ˜¾ç¤ºã€‚ |
| `patch_artist` | æ˜¯å¦å…è®¸å¡«å……é¢œè‰²ã€‚ |
| `tick_labels` | æ¯ä¸ªç®±å­çš„æ ‡ç­¾ã€‚ |
| `showfliers` | æ˜¯å¦æ˜¾ç¤ºå¼‚å¸¸å€¼ã€‚ |
| `widths` | ç®±ä½“å®½åº¦ã€‚ |

```python
np.random.seed(10)
data1 = np.random.normal(100, 10, 200)
data2 = np.random.normal(80, 30, 200)
data3 = np.random.normal(90, 20, 200)
data_to_plot = [data1, data2, data3]

fig, ax = plt.subplots()
bp = ax.boxplot(
    data_to_plot,
    patch_artist=True,
    tick_labels=["Aäº§å“", "Bäº§å“", "Cäº§å“"],
    showfliers=True,
    widths=0.6
)

colors = ["#99ff99", "#66b3ff", "#ff9999"]
for patch, color in zip(bp["boxes"], colors):
    patch.set_facecolor(color)

ax.set_title("ä¸åŒäº§å“é”€å”®é¢åˆ†å¸ƒå¯¹æ¯”")
ax.set_ylabel("é”€å”®é¢")
ax.yaxis.grid(True)

plt.show()
```

### å››ã€å®šåˆ¶ä½ çš„å›¾è¡¨ï¼šè®©ä¿¡æ¯æ›´æ¸…æ™°


ä¸ºå›¾è¡¨æ·»åŠ æ ‡é¢˜ã€æ ‡ç­¾ã€å›¾ä¾‹ã€ç½‘æ ¼æˆ–æ–‡æœ¬æ³¨è§£ï¼Œå¯ä»¥æ˜¾è‘—æå‡å¯è¯»æ€§ã€‚ä¸‹é¢é€šè¿‡å‡ ä¸ªç¤ºä¾‹æ¼”ç¤ºå¸¸è§çš„å®šåˆ¶æŠ€å·§ã€‚

#### 1. æ·»åŠ æ ‡é¢˜ã€æ ‡ç­¾å’Œå›¾ä¾‹


é€šè¿‡ `ax.set_title`ã€`ax.set_xlabel`ã€`ax.set_ylabel` ä»¥åŠ `ax.legend()` å¯ä»¥è®©å›¾è¡¨è¡¨è¾¾æ›´æ˜ç¡®ã€‚

```python
x = np.linspace(0, 2 * np.pi, 100)


y1 = np.sin(x)


y2 = np.cos(x)




fig, ax = plt.subplots()


ax.plot(x, y1, label="sin(x)")


ax.plot(x, y2, label="cos(x)", linestyle="--")


ax.set_title("æ­£å¼¦ä¸ä½™å¼¦å‡½æ•°å›¾åƒ")


ax.set_xlabel("X è½´ (å¼§åº¦)")


ax.set_ylabel("Y è½´ (å€¼)")


ax.legend()


plt.show()
```

#### 2. è°ƒæ•´æ ·å¼ã€é¢œè‰²å’Œç½‘æ ¼


å¯ä»¥åœ¨ `plot` ä¸­ç›´æ¥æŒ‡å®šé¢œè‰²ã€çº¿å‹ä¸æ ‡è®°ï¼Œå¹¶é€šè¿‡ `ax.grid`ã€`ax.set_xlim` ç­‰æ–¹æ³•å¾®è°ƒå›¾è¡¨ç»†èŠ‚ã€‚

```python
fig, ax = plt.subplots()


ax.plot(x, y1, label="sin(x)", color="blue", linestyle="-", marker="o", markersize=2)


ax.plot(x, y2, label="cos(x)", color="red", linestyle=":")


ax.grid(True, linestyle=":", alpha=0.6)


ax.set_xlim(0, 2 * np.pi)


ax.set_ylim(-1.5, 1.5)


ax.set_title("å®šåˆ¶åŒ–æ ·å¼çš„å›¾è¡¨")


ax.legend()


plt.show()
```

#### 3. æ·»åŠ æ–‡æœ¬æ³¨è§£å’Œæ—‹è½¬åˆ»åº¦


å½“ç±»åˆ«æ ‡ç­¾è¿‡é•¿æˆ–éœ€è¦çªå‡ºæŸäº›ç‚¹æ—¶ï¼Œå¯ä»¥ç»“åˆ `ax.tick_params` å’Œ `ax.text` æ¥æ ‡æ³¨ç»†èŠ‚ã€‚

```python
categories = ["éå¸¸ä¸æ»¡æ„", "ä¸æ»¡æ„", "ä¸€èˆ¬", "æ»¡æ„", "éå¸¸æ»¡æ„"]


values = [5, 25, 50, 120, 200]




fig, ax = plt.subplots(figsize=(8, 5))


bars = ax.bar(categories, values, color="#66b3ff")


ax.tick_params(axis="x", rotation=45)




for bar in bars:


    yval = bar.get_height()


    ax.text(bar.get_x() + bar.get_width() / 2.0, yval, int(yval), va="bottom", ha="center")




ax.set_title("å®¢æˆ·æ»¡æ„åº¦è°ƒæŸ¥ç»“æœ")


ax.set_ylabel("æŠ•ç¥¨æ•°")


fig.tight_layout()


plt.show()
```

### äº”ã€å¤šå­å›¾å¸ƒå±€ (`subplots`)

å½“ä½ æƒ³å°†å¤šä¸ªç›¸å…³çš„å›¾è¡¨å¹¶æ’æ¯”è¾ƒæ—¶ï¼Œ`plt.subplots()` æä¾›äº†æœ€ç›´æ¥çš„å…¥å£ï¼Œå¯ä»¥ä¸€æ¬¡æ€§åˆ›å»ºå¸¦æœ‰å¤šä¸ªåæ ‡è½´çš„ç”»å¸ƒã€‚



| å‚æ•° | è¯´æ˜ |

| --- | --- |

| `nrows`, `ncols` | å­å›¾ç½‘æ ¼çš„è¡Œåˆ—æ•°ã€‚ |

| `figsize` | æ•´ä¸ªç”»å¸ƒçš„å®½å’Œé«˜ï¼ˆå•ä½è‹±å¯¸ï¼‰ã€‚ |

| `sharex`, `sharey` | æ˜¯å¦åœ¨å¤šä¸ªå­å›¾ä¹‹é—´å…±äº«åæ ‡è½´ã€‚ |

| `squeeze` | æ§åˆ¶è¿”å›çš„ `axes` ç»´åº¦ï¼›è®¾ä¸º `False` æ—¶æ€»æ˜¯è¿”å›äºŒç»´æ•°ç»„ã€‚ |

| `gridspec_kw` | ä¼ é€’ç»™åº•å±‚ `GridSpec` çš„å­—å…¸ï¼Œç”¨äºæ›´ç²¾ç»†åœ°æ§åˆ¶å¸ƒå±€ã€‚ |



ä¸‹é¢çš„ä¾‹å­å±•ç¤ºäº†å¦‚ä½•åœ¨ä¸€ä¸ª 2Ã—2 çš„å­å›¾ç½‘æ ¼ä¸­ç»˜åˆ¶å››ç§ä¸åŒç±»å‹çš„å›¾è¡¨ã€‚

```python
fig, axes = plt.subplots(

    nrows=2,

    ncols=2,

    figsize=(10, 8),

    constrained_layout=True,

    squeeze=True

)



# `axes` æ˜¯ä¸€ä¸ª 2x2 çš„æ•°ç»„ï¼Œé€ä¸ªå­å›¾è¿›è¡Œç»˜åˆ¶

axes[0, 0].plot(np.sin(np.linspace(0, 10, 100)))

axes[0, 0].set_title("æŠ˜çº¿å›¾")



axes[0, 1].bar(["A", "B", "C"], [3, 5, 2])

axes[0, 1].set_title("æŸ±çŠ¶å›¾")



axes[1, 0].hist(np.random.randn(500), bins=20)

axes[1, 0].set_title("ç›´æ–¹å›¾")



axes[1, 1].scatter(np.random.rand(50), np.random.rand(50))

axes[1, 1].set_title("æ•£ç‚¹å›¾")



plt.show()
```

```python
output_path = Path("images/matplotlib")

output_path.mkdir(parents=True, exist_ok=True)



fig, ax = plt.subplots(figsize=(6, 4))

x = np.linspace(0, 2 * np.pi, 100)

y1 = np.sin(x)

ax.plot(x, y1, label="sin(x)")

ax.set_title("ä¿å­˜è¿™å¼ å›¾")

ax.legend()



save_path = output_path / "figure_4_12_saved_plot.png"

fig.savefig(

    save_path,

    dpi=300,

    bbox_inches="tight",

    facecolor="white",

    transparent=False

)

plt.show()



print(f"å›¾è¡¨å·²ä¿å­˜ä¸º {save_path}")
```
